# cleaning

```python
import re

# เปลี่ยนเป็น lower case
df["clean_posts"] = df["posts"].str.lower()

# เปลี่ยน ||| เป็นช่องว่าง
df["clean_posts"] = df["clean_posts"].str.replace(
    re.compile(r"\|\|\|"), " "
)

df["clean_posts"] = df["clean_posts"].str.replace(
    re.compile(r"https?:\/\/(www)?.?([A-Za-z_0-9-]+)([\S])*"), ""
)

# ลบ email
df["clean_posts"] = df["clean_posts"].str.replace(
    re.compile(r"\S+@\S+"), ""
)
# ลบเครื่องหมายต่างๆ
df["clean_posts"] = df["clean_posts"].str.replace(
    re.compile(r"[^a-z\s]"), " "
)


# ลบอันที่มี MBTIs ในโพส์
mbti = df["type"].unique()
for type_word in mbti:
    df["clean_posts"] = df["clean_posts"].str.replace(
        type_word.lower(), ""
    )

df.head()
```

```python
#lemmitizing
import time
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download("stopwords")
nltk.download('wordnet')

t = time.time()

lemmatizer = WordNetLemmatizer()

df["clean_posts"] = df["clean_posts"].apply(
    lambda x: " ".join(
        [
            lemmatizer.lemmatize(word)
            for word in x.split(" ")
            if word not in stopwords.words("english")
        ]
    )
)

print(f"Lemmitizing Time: {time.time() - t} seconds")
```

```python
# ลบข้อความยาวๆ
df["clean_posts"] = df["clean_posts"].str.replace(
    re.compile(r"\b\w{1,2}\b"), ""
)

df.head()
```

# Sentiments Analysis

```python
#วิเคราห์
import time
t = time.time()

# sentiment scoring

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

nlp_sentiment_score = []

for post in df["clean_posts"]:
    score = analyzer.polarity_scores(post)
    nlp_sentiment_score.append(score)

print(f"Sentiment Scoring Time: {time.time() - t:.2f} seconds")
```

```python
# แยก sentiment
df["compound_sentiment"] = [
    score["compound"] for score in nlp_sentiment_score
]
df["pos_sentiment"] = [score["pos"] for score in nlp_sentiment_score]
df["neg_sentiment"] = [score["neg"] for score in nlp_sentiment_score]
df["neu_sentiment"] = [score["neu"] for score in nlp_sentiment_score]

df.head()
```

# POS Tagging

```python
# pos tagging คำ
t = time.time()

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

df["tagged_words"] = df["clean_posts"].apply(
    lambda x: [nltk.pos_tag(word_tokenize(line)) for line in x]
)

print(f"POS Tagging Time: {time.time() - t} seconds")
```

```python
# สร้าง list ของ POS tags
tag_set = set()

for i, data in df["tagged_words"].iteritems():
    for tup in data[0]:
        tag_set.add(tup[1])

tag_list = list(tag_set)

# คำนวน mean และ standard deviation
t = time.time()

def pos_cat(x, tag):
    return [len([y for y in line if y[1] == tag]) for line in x]


for col in tag_list:
    df["POS_" + col + "_mean"] = df["tagged_words"].apply(
        lambda x: np.mean(pos_cat(x, col))
    )
    df["POS_" + col + "_std"] = df["tagged_words"].apply(
        lambda x: np.std(pos_cat(x, col))
    )
    
# จัดกลุ่ม pos tags ตาม stanford list
tags_dict = {
    "ADJ": ["JJ", "JJR", "JJS"],
    "ADP": ["EX", "TO"],
    "ADV": ["RB", "RBR", "RBS", "WRB"],
    "CONJ": ["CC", "IN"],
    "DET": ["DT", "PDT", "WDT"],
    "NOUN": ["NN", "NNS", "NNP", "NNPS"],
    "NUM": ["CD"],
    "PRT": ["RP"],
    "PRON": ["PRP", "PRP$", "WP", "WP$"],
    "VERB": ["MD", "VB", "VBD", "VBG", "VBN", "VBP", "VBZ"],
    ".": ["#", "$", "''", "(", ")", ",", ".", ":"],
    "X": ["FW", "LS", "UH"],
}

# Stanford POS tag stats
t = time.time()

def stanford_tag(x, tag):
    tags_list = [len([y for y in line if y[1] in tags_dict[col]]) for line in x]
    return tags_list

for col in tags_dict.keys():
    df[col + "_avg"] = df["tagged_words"].apply(
        lambda x: np.median(stanford_tag(x, col))
    )

df.head()    
```

# COUNTS

```python
# ดู Question,Exclamation,Colon,Emoji 

def unique_words(s):
    unique = set(s.split(" "))
    return len(unique) / 50


def emojis(post): # ไม่นับ emojis ที่ทำจาก symbols, only :word:
    emoji_count = 0
    words = post.split()
    for e in words:
        if "http" not in e:
            if e.count(":") == 2:
                emoji_count += 1
    return emoji_count / 50


def colons(post): # นับ colons ที่ใช้ใน emojis
    colon_count = 0
    words = post.split()
    for e in words:
        if "http" not in e:
            colon_count += e.count(":")
    return colon_count / 50

df["qm"] = df["posts"].apply(lambda s: s.count("?") / 50)
df["em"] = df["posts"].apply(lambda s: s.count("!") / 50)
df["colons"] = df["posts"].apply(colons)
df["emojis"] = df["posts"].apply(emojis)
```

```python
# Word Count

df["word_count"] = df["posts"].apply(
    lambda s: (s.count(" ") + 1) / 50
)
df["unique_words"] = df["posts"].apply(unique_words)


# Word Stats

t = time.time()

# df2["avg_word_ct"] = df2["word_count"].apply(lambda s: s / 50)

df["post_length_var"] = df["posts"].apply(
    lambda x: np.var([len(post.split()) for post in x.split("|||")])
)

print(f"Time Taken: {time.time() - t}")
```

```python
# Upper Case Count

df["upper"] = df["posts"].apply(
    lambda x: len([x for x in x.split() if x.isupper()]) / 50
)


#Link Count

df["link_count"] = df["posts"].apply(
    lambda s: s.count("http") / 50
)


#Ellipses Count

ellipses_count = [
    len(re.findall(r"\.\.\.\ ", posts)) / 50 for posts in df["posts"]
]
df["ellipses"] = ellipses_count


#Image Count
df["img_count"] = [
    len(re.findall(r"(\.jpg)|(\.jpeg)|(\.gif)|(\.png)", post)) / 50
    for post in df["posts"]
]


df.head()
```
